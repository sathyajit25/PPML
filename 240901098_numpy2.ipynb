```python
import numpy as np
#joining two array
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
arr=np.concatenate((arr1,arr2))
print(arr)
 
```

```python
#horizondal line
arr3=np.hstack((arr1,arr2))
print(arr3)
```

```python
arr4=np.vstack((arr1,arr2))
print(arr3)
```

```python
#depth join
arr5=np.dstack((arr1,arr2))
print(arr3)
```

```python
#splitting of array
arr=np.array([1,2,3,4,5,6])
arr6=np.array_split(arr,3)
print(arr)
print(arr6)
```
